```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from xgboost import XGBClassifier

from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/heart.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

#Preprocessing


```python
x=df.iloc[:, :-1]
y=df.iloc[:, -1]
```

```python
x
```

```python
le_model = LabelEncoder()

x['Sex'] = le_model.fit_transform(x['Sex'])
x['ChestPainType'] = le_model.fit_transform(x['ChestPainType'])
x['RestingECG'] = le_model.fit_transform(x['RestingECG'])
x['ExerciseAngina'] = le_model.fit_transform(x['ExerciseAngina'])
x['ST_Slope'] = le_model.fit_transform(x['ST_Slope'])
```

```python
StandardScalerModel = StandardScaler()
x = StandardScalerModel.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0, shuffle=True)
```

#Create model


```python
model = XGBClassifier(n_estimators=1000, learning_rate=0.01)
model.fit(x_train, y_train, verbose=True, early_stopping_rounds=5, eval_set=[(x_test, y_test)])
```

```python
print(model.score(x_train, y_train))
```

```python
print(model.score(x_test, y_test))
```
